## 4. XGBoost model
In this part, the XGBoost model was implemented. XGBoost model was chosen over other boosting models (like AdaBoost or Gradient Boosting) because it better handels inbalaced dataset by combining strong regularization and flexible loss weighting. Feature groups created earlier (for the Random Forest model) were used also for this model. In the next part of script, hyperparameter tuning will be performed and the best parameters for each feature group will be selected in the cross-validation process. Finally, models for all groups will be tested on the validation set, and the final model with the best result will be selected and tested on test set.

### Hyperparametes Tunning for each group of variables
The Grid Search Cross Validation method was used to tune the hyperparameters for the XGBoost model. Hyperparameters we tune:
* **max_depth** - Maximum depth of a single tree
* **learning_rate** - Controls step size
* **n_estimators** - Number of boosting rounds (trees in the ensemble)
* **gamma** - Minimum loss reduction required to make a split (tree pruning parameter)
* **reg_lambda** - L2 regularization on leaf weights
* **scale_pos_weight** - Weight for positive class in imbalanced classification. It is set manually to 15 and we don't tune it becuse we know that negative:positive ratio in our dataset is close to 15:1

```python
param_grid = {
    'max_depth': [3, 4, 5],
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [300, 600],
    'gamma': [0, 0.25],
    'reg_lambda': [1, 10],
    'scale_pos_weight': [15]  
}

optimal_params = GridSearchCV(
    estimator=xgb.XGBClassifier(
        objective="binary:logistic", # It specifies loss function for classification problem
        eval_metric="logloss", # This metric penalizes wrong predictions more strongly if the predicted probability is confident but wrong
        subsample=0.8, # It reduces overfitting
        colsample_bytree=0.5, # Fraction of features randomly selected for each tree (also reduces overfitting)
        random_state=42,
        n_jobs=-1
    ),
    param_grid=param_grid,
    scoring="balanced_accuracy",
    verbose=2,
    n_jobs=-1,          
    cv=5
)
```

### Now as we chose best hyperparameters for each feature group we can choose best model using validation dataset

```python
feature_groups = {
    "benchmark": benchmark,
    "benchmark_input": benchmark_input,
    "var": var,
    "mi_features_15": mi_features_15,
    "mi_features_25": mi_features_25,
    "corr_features": corr_features,
    "pi_var_prim": pi_var_prim,
    "relief3": relief3,
    "br_features": br_features,
}

xgb_params = {
    "benchmark": {'gamma': 0.25, 'learning_rate': 0.01, 'max_depth': 5, 'n_estimators': 300, 'reg_lambda': 10,},
    "benchmark_input": {'gamma': 0.25, 'learning_rate': 0.01, 'max_depth': 3, 'n_estimators': 600, 'reg_lambda': 10,},
    "var": {'gamma': 0, 'learning_rate': 0.01, 'max_depth': 3, 'n_estimators': 300, 'reg_lambda': 1,},
    "mi_features_15": {'gamma': 0, 'learning_rate': 0.01, 'max_depth': 3, 'n_estimators': 300, 'reg_lambda': 1,},
    "mi_features_25": {'gamma': 0, 'learning_rate': 0.01, 'max_depth': 3, 'n_estimators': 300, 'reg_lambda': 1,},
    "corr_features": {'gamma': 0.25, 'learning_rate': 0.01, 'max_depth': 3, 'n_estimators': 600, 'reg_lambda': 10,},
    "pi_var_prim": {'gamma': 0.25, 'learning_rate': 0.01, 'max_depth': 4, 'n_estimators': 600, 'reg_lambda': 10,},
    "relief3": {'gamma': 0, 'learning_rate': 0.01, 'max_depth': 3, 'n_estimators': 300, 'reg_lambda': 1,},
    "br_features": {'gamma': 0.25, 'learning_rate': 0.01, 'max_depth': 4, 'n_estimators': 600, 'reg_lambda': 10,},}

best_group = None
best_score = 0
results = {}

for name, features in feature_groups.items():
    params = xgb_params.get(name, {})  
    model = xgb.XGBClassifier(**params,
                             scale_pos_weight = 15,
                             random_state = 42,
                             objective= "binary:logistic",
                             n_jobs = -1)

    scores = cross_val_score(
        model,
        X_train[features],
        y_train,
        cv=5,
        scoring="balanced_accuracy"
    )
    mean_score = scores.mean()
    results[name] = mean_score

    print(f"Feature group: {name}, Balanced Accuracy (CV): {mean_score:.4f}, Params: {params}")

    if mean_score > best_score:
        best_score = mean_score
        best_group = name
        best_model = model


print(f"\nBest group: {best_group} with CV score: {best_score:.4f}")
```

#### Fitting and saving best model
Best feature group: br_features, parameters: {'gamma': 0.25, 'learning_rate': 0.01, 'max_depth': 4, 'n_estimators': 600, 'reg_lambda': 10}

```python
best_model = xgb.XGBClassifier(
                            gamma = 0.25,
                            learning_rate = 0.01,
                            max_depth = 4,
                            n_estimators = 600,
                            reg_lambda = 10,
                            scale_pos_weight = 15,
                            random_state = 42,
                            objective= "binary:logistic",
                            n_jobs = -1)
```

```python
best_model.fit(df_res.loc[:, br_features].values, df_res.loc[:, "claim_status"].values.ravel())
```

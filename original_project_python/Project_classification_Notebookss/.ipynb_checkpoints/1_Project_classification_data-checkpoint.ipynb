The project concerns the prediction of insurance claims reported in relation to various vehicles covered by insurance policies.

The data comes from Kaggle from a dataset called *Insurance Claims Dataset*, https://www.kaggle.com/datasets/litvinenko630/insurance-claims. This dataset contains nearly 60,000 observations, each observation refers to a single policy and contains detailed attributes related to the vehicle (such as engine power, fuel type, and the presence of various safety systems), as well as characteristics of the policy and the policyholder (such as policy duration, driver age, and region of residence).

The dataset addresses a key problem in the field of insurance risk assessment and fraud detection: predicting the likelihood of a claim being filed. Understanding which factors contribute to a higher likelihood of a claim allows insurance companies to price premiums more accurately, develop fairer risk models, and improve overall profitability. In addition, such predictive modeling can help insurers promote safer driving behaviors and identify high-risk policies early on. 

**The aim of the project is to forecast insurance claims as accurately as possible using selected machine learning models.** Three models were selected:
* Random Forest
* XGBoost
* Ensembling of three models with the Stacking selection method.

The project has the following structure: firt, data analysis and cleaning, then feature engineering and feature selection, followed by training each model and tuning hyperparameters, and finally testing the models on a test set and comparing the results. To this end, the project has been divided into 5 appropriate modules (Jupyter Notebooks) to increase its readability and simplify code handling.


## 1. Data cleaning and data analysis
In this section, a preliminary analysis of the data (distribution of variables, correlations between them) will be performed, and the dataset will be cleaned of unnecessary variables and missing values. The dataset will also be adapted for further analysis—division into training and test sets (while maintaining reproducibility of research) and encoding of categorical variables. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
import scipy.stats as stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from pathlib import Path

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
```

```python
df = pd.read_csv('C:\\Users\\kubas\\OneDrive\\Documents\\ML2\\ProjectClassification\\Insurance_claims_data.csv')
```

### Splitting data into train and test datasets

```python
df_train, df_test = train_test_split(df, test_size=0.2, random_state=123) # Random state assures reproducibility 
```

```python
print("\nTRAIN:")
display(df_train.head())
print("SHAPE:",df_train.shape)

print("\nTEST:")
display(df_test.head())
print("SHAPE:",df_test.shape)
```

### Dataset cleaning

```python
df_train.info()
```

##### Checking for missing values

```python
df_train.isna().sum()/(df.size/14)*100
```

There are no missing values in our dataset.

##### We want to deleate not significant features like policy_id which has different value for each observation, so it doesn't add any more information to the dataset

```python
df_train = df_train.drop(['policy_id', 'region_code', 'engine_type'], axis=1)
df_test = df_test.drop(['policy_id', 'region_code', 'engine_type'], axis=1)
```

#### Encoding object type variables:
For that we use one hot encoding

```python
df_new = df_train.copy()
df_test_new = df_test.copy()
```

```python
categorical_columns = df_new.select_dtypes(include=['object']).columns

for i in range(0,len(categorical_columns)):
    
    y = pd.get_dummies(df_new[categorical_columns[i]], prefix=categorical_columns[i])

    df_new=pd.concat([df_new,pd.get_dummies(df_new[categorical_columns[i]],prefix=categorical_columns[i])],axis=1).drop([categorical_columns[i]],axis=1)

print("Frequency of data type")
df_new.dtypes.value_counts()
```

```python
categorical_columns = df_test_new.select_dtypes(include=['object']).columns

for i in range(0,len(categorical_columns)):
    
    y = pd.get_dummies(df_test_new[categorical_columns[i]], prefix=categorical_columns[i])

    df_test_new=pd.concat([df_test_new,pd.get_dummies(df_test_new[categorical_columns[i]],prefix=categorical_columns[i])],axis=1).drop([categorical_columns[i]],axis=1)

print("Frequency of data type")
df_test_new.dtypes.value_counts()
```
